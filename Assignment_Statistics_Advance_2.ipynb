1  What is hypothesis testing in statistics?

Ans   Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample of data. It involves testing an assumption (called a hypothesis) about a population parameter.

2  What is the null hypothesis, and how does it differ from the alternative hypothesis?


Ans  The null hypothesis (H₀) and the alternative hypothesis (H₁ or Ha) are two competing statements in hypothesis testing, and they serve different purposes in statistical analysis

Differences:
H₀ represents no change or no effect, while H₁ represents the presence of a change or effect.

H₀ is typically assumed to be true at the start of the hypothesis test, and it can only be rejected if there is sufficient evidence in favor of H₁.

The alternative hypothesis (H₁) is what you aim to support or prove through your data

3  What is the significance level in hypothesis testing, and why is it important?


Ans  The significance level in hypothesis testing, denoted by α (alpha), is the threshold used to decide whether to reject the null hypothesis (H₀). It represents the maximum probability of making a Type I error, which occurs when you wrongly reject a true null hypothesis.

important
Controls error risk


Guides decision-making


4    What does a P-value represent in hypothesis testing?

Ans   A p-value in hypothesis testing represents the probability of obtaining a result as extreme as (or more extreme than) the observed result, assuming the null hypothesis (H₀) is true.

5  How do you interpret the P-value in hypothesis testing?


Ans  Steps to Interpret the P-value:
Compare the p-value to the significance level (α):

If p-value ≤ α:

Reject the null hypothesis (H₀).

This suggests that the sample data provides enough evidence to support the alternative hypothesis (H₁).
If p-value > α:

Fail to reject the null hypothesis (H₀).

This suggests that the sample data does not provide enough evidence to support the alternative hypothesis (H₁).

The result is not statistically significant.

6   What are Type 1 and Type 2 errors in hypothesis testing?

Ans   Type I Error (False Positive):
Definition: Rejecting the null hypothesis (H₀) when it is actually true.

You say there's an effect or difference when there isn't.

Probability of this error: α (significance level), often set at 0.05

Type II Error (False Negative):
Definition: A Type II error occurs when you fail to reject the null hypothesis (H₀) when the alternative hypothesis (H₁) is actually true.

Interpretation: You fail to detect an effect or difference that actually exists.

Also known as: False negative.

Probability: The probability of committing a Type II error is denoted by β.

9   How do you calculate the Z-score, and what does it represent in hypothesis testing?


Ans    Population Z-Score
When the population mean (μ) and population standard deviation (σ) are known:

Z = (X – μ) / σ   

X: Observed data point

μ: Population mean

σ: Population standard deviation


Sample Z-Score
When using sample data to estimate the population parameters:

Z = (X – x̄) / (s / √n)

X: Observed data point

x̄: Sample mean

s: Sample standard deviation

n: Sample size

```python
#2  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

def bernoulli_pmf(p):
    """
    Calculate and plot the PMF of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 <= p <= 1)
    """
    # Values of x (0 and 1)
    x = np.array([0, 1])

    # Calculate PMF for each x
    pmf_values = bernoulli.pmf(x, p)

    # Display the PMF values
    for xi, pmf in zip(x, pmf_values):
        print(f"P(X = {xi}) = {pmf}")

    # Plotting the PMF
    plt.figure(figsize=(8, 6))
    plt.stem(x, pmf_values, basefmt=" ", use_line_collection=True)
    plt.title(f'Bernoulli Distribution PMF (p = {p})')
    plt.xlabel('x')
    plt.ylabel('P(X = x)')
    plt.xticks(x)
    plt.ylim(0, 1)
    plt.grid(True)
    plt.show()

# Example usage
bernoulli_pmf(0.3)
```

```python
#9  Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
        data (array-like): Input data for which Z-scores are to be calculated.

    Returns:
        np.ndarray: Z-scores of the input data.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(data):
    """
    Plot the Z-scores of a dataset.

    Parameters:
        data (array-like): Input data to be visualized.
    """
    z_scores = calculate_z_scores(data)

    # Plotting the Z-scores
    plt.figure(figsize=(10, 6))
    sns.histplot(z_scores, kde=True, bins=30, color='skyblue', edgecolor='black')
    plt.axvline(x=0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title("Z-Scores Distribution")
    plt.xlabel("Z-Score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data = np.random.normal(loc=50, scale=10, size=1000)  # Generating random data
plot_z_scores(data)
```

```python
#15  Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
import scipy.stats as stats

# Sample data
sample_data = np.array([205, 198, 210, 190, 215, 205, 200, 192, 198, 205,
                        198, 202, 208, 200, 205, 198, 205, 210, 192, 205])

# Population parameters
population_mean = 200
population_std_dev = 5

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Z-test statistic
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# P-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    result = "Reject the null hypothesis."
else:
    result = "Fail to reject the null hypothesis."

# Output
print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
print(result)
```

```python
#16 Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

# Sample data
data = [12, 15, 14, 10, 13, 17, 14, 15, 16, 14]

# Sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
sample_size = len(data)

# Confidence level
confidence_level = 0.95

# Calculate the standard error
standard_error = sample_std / np.sqrt(sample_size)

# Calculate the t-statistic for the desired confidence level
t_statistic = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

# Calculate the margin of error
margin_of_error = t_statistic * standard_error

# Calculate the confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"The {confidence_level*100}% confidence interval for the population mean is: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#17  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats

# Step 1: Generate random data from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=5, size=1000)  # Mean=50, Std=5, Sample size=1000

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
sample_size = len(data)

# Step 3: Define confidence level
confidence_level = 0.95

# Step 4: Calculate the critical value (z-score) for the confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Step 5: Calculate the margin of error
margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

# Step 6: Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Critical Z-score: {z_score:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"{int(confidence_level * 100)}% Confidence Interval: {confidence_interval}")
```

```python
#Practial 2
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
import scipy.stats as stats

# Sample data
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Mean=100, Std=15, n=50

# Population parameters
population_mean = 105
population_std = 15

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
standard_error = population_std / np.sqrt(sample_size)

# Z-test statistic
z_statistic = (sample_mean - population_mean) / standard_error

# P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu_0 = 100  # Population mean
sigma = 15  # Population standard deviation
n = 30      # Sample size
alpha = 0.05  # Significance level

# Sample data
np.random.seed(0)
sample = np.random.normal(loc=102, scale=sigma, size=n)
sample_mean = np.mean(sample)

# Calculate Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Critical values for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Plotting
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label=f'Reject H₀ if Z < {-z_critical:.2f}')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label=f'Reject H₀ if Z > {z_critical:.2f}')
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')
plt.title('Two-Tailed Z-Test Visualization')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Output Z-statistic and critical values
print(f"Z-statistic: {z_stat:.2f}")
print(f"Critical Z-values: ±{z_critical:.2f}")
```

```python
#5  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type_errors(population_mean, sample_size, alpha=0.05, effect_size=5, num_simulations=1000):
    """
    Simulates hypothesis testing to visualize Type I and Type II errors.

    Parameters:
    - population_mean: Mean of the population under the null hypothesis.
    - sample_size: Number of samples in each simulation.
    - alpha: Significance level for the hypothesis test.
    - effect_size: Difference between the population mean and the true mean under the alternative hypothesis.
    - num_simulations: Number of simulations to run.

    Returns:
    - None
    """
    # Initialize counters for errors
    type_i_errors = 0
    type_ii_errors = 0

    # Arrays to store p-values and rejection decisions
    p_values = []
    decisions = []

    # Run simulations
    for _ in range(num_simulations):
        # Simulate data under the null hypothesis
        null_data = np.random.normal(loc=population_mean, scale=10, size=sample_size)
        # Simulate data under the alternative hypothesis
        alt_data = np.random.normal(loc=population_mean + effect_size, scale=10, size=sample_size)

        # Perform t-test for both datasets
        t_stat_null, p_value_null = stats.ttest_1samp(null_data, population_mean)
        t_stat_alt, p_value_alt = stats.ttest_1samp(alt_data, population_mean)

        # Record p-values
        p_values.append((p_value_null, p_value_alt))

        # Decision based on null hypothesis
        reject_null = p_value_null < alpha
        decisions.append((reject_null, p_value_alt < alpha))

        # Count errors
        if reject_null:
            type_i_errors += 1
        if not reject_null and p_value_alt < alpha:
            type_ii_errors += 1

    # Calculate error rates
    type_i_error_rate = type_i_errors / num_simulations
    type_ii_error_rate = type_ii_errors / num_simulations

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.hist([p[0] for p in p_values], bins=30, alpha=0.5, label='Null Hypothesis (H0)', color='blue')
    plt.hist([p[1] for p in p_values], bins=30, alpha=0.5, label='Alternative Hypothesis (H1)', color='red')
    plt.axvline(x=alpha, color='black', linestyle='--', label=f'Significance Level (α={alpha})')
    plt.title('Distribution of p-values under Null and Alternative Hypotheses')
    plt.xlabel('p-value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    print(f"Type I Error Rate: {type_i_error_rate:.4f}")
    print(f"Type II Error Rate: {type_ii_error_rate:.4f}")

# Example usage
visualize_type_errors(population_mean=100, sample_size=30, alpha=0.05, effect_size=5, num_simulations=1000)
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: List or array of sample data.
    - confidence_level: Desired confidence level (default is 0.95 for 95%).

    Returns:
    - tuple: Lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(data)

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Critical value from the t-distribution
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Margin of error
    margin_of_error = t_critical * standard_error

    # Confidence interval
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return ci_lower, ci_upper
```

```python
#11   Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, false_positive_rate, prevalence):
    """
    Calculate the posterior probability using Bayes' Theorem.

    :param prior: Prior probability of the disease (P(Disease))
    :param likelihood: Likelihood of a positive test given the disease (P(Pos|Disease))
    :param false_positive_rate: Probability of a positive test given no disease (P(Pos|No Disease))
    :param prevalence: Prevalence of the disease in the population (P(Disease))
    :return: Posterior probability (P(Disease|Pos))
    """
    # Calculate the marginal likelihood (P(Pos))
    marginal_likelihood = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Apply Bayes' Theorem
    posterior = (likelihood * prior) / marginal_likelihood
    return posterior

# Given values
prior = 0.01  # P(Disease)
likelihood = 0.95  # P(Pos|Disease)
false_positive_rate = 0.05  # P(Pos|No Disease)
prevalence = 0.01  # P(Disease)

# Calculate posterior probability
posterior = bayes_theorem(prior, likelihood, false_positive_rate, prevalence)
print(f"Posterior Probability: {posterior:.4f}")
```

```python
#13  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np
from scipy.stats import chisquare

# Example observed data (e.g., counts of categories)
observed = [25, 30, 45, 50]

# Calculate total observations
total_observed = sum(observed)

# Calculate number of categories
num_categories = len(observed)

# Calculate expected frequencies assuming uniform distribution
expected = [total_observed / num_categories] * num_categories

# Perform Chi-square test
chi2_stat, p_value = chisquare(observed, f_exp=expected)

# Display results
print(f"Observed frequencies: {observed}")
print(f"Expected frequencies: {expected}")
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies significantly differ from the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies do not significantly differ from the expected frequencies.")
```
